<h1>Breakdown of Greenhouse Gas Emissions Trends for a Sample of Countries Between the Years 1990 to 2016</h1>
<h2>Introduction</h2>
<p>I attainded this dataset from data.un.org</p>
<h2>Metadata</h2>
<p>The data is divided into three columns: 'Country or area', 'year', and 'value' representing kilotonnes of greenhouse gas. The countries or areas represented in this dataset are the Czech Republic, Denmark, the European Union, Finland, Japan, Latvia, the Netherlands, Portugal, and Switzerland.</p>
<p>The script below is the script I used to extract and parse the data.</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
sb.set_context('paper')

#Import as a pandas data frame
#Greenhouse Gas (GHGs) Emissions, including Indirect CO2, without LULUCF, in kilotonne CO2 equivalent
df = pd.read_csv("UNdata_Export_20190128_224045635.csv")

#Get Column Names
print('Columns represented:\n,',list(df))

#Lets make this a little easier to work with
df = df.rename(columns = {'Country or Area':'Country','Value' : 'Kilotonnes'})

#List unique entries in the 'country' tab
print('Countries represented:\n',df['Country'].unique())

#List the years represented
print('Years represented:\n',df['Year'].unique())
```

```python
#Lets make a function!
def subPlotEmissions(df, nX, nY):
    n = 1  #Map index
    
    plt.figure(figsize=(14,8))
    for country in df['Country'].unique():
        
        #Dataframe for general case
        country_df = df.loc[lambda df: df['Country'] == country] 
        
        #Generate Subplot
        plt.subplot(nX,nY,n)
        plt.plot(country_df['Year'], country_df['Kilotonnes'])
        
        #Plot lables
        plt.title(country)
        plt.xlabel('Year')
        plt.ylabel('Kilotonnes')
        
        #Plot properties
        plt.tight_layout()
        n += 1
        
subPlotEmissions(df, 3, 3)
plt.show()
```

<h2>Analysis</h2>
<p>As we can see by the data, most of the sample regions have experienced a reduction in greenhouse gas emissions since 1990 with the exception of Japan and Portugal. All sample countries have a local peak in greenhouse gas emissions in the year 1996 wth the exception of Portugal, who's data shows a local peak in 1995. By magnitude, the European Union has the highest greenhouse gas emissions out of all sample regions, but it is important to note that many of the regions, namely the Netherlands, Portugal, Finland, Latvia, Denmark, and the Czech Republic are member states of the European Union.</p>
